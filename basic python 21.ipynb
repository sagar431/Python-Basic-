1. Add the current date to the text file today.txt as a string.

```python
from datetime import date

today = date.today()

with open('today.txt', 'w') as file:
    file.write(str(today))
```

2. Read the text file today.txt into the string today_string

```python
with open('today.txt', 'r') as file:
    today_string = file.read()

print(today_string)  # Output: 2022-05-05
```

3. Parse the date from today_string.

```python
from datetime import datetime

parsed_date = datetime.strptime(today_string, '%Y-%m-%d').date()
print(parsed_date)  # Output: 2022-05-05
```

7. Create a date object of your day of birth.

```python
from datetime import date

my_birthdate = date(1990, 10, 20)
print(my_birthdate)  # Output: 1990-10-20
```

8. What day of the week was your day of birth?

```python
from datetime import date

my_birthdate = date(1990, 10, 20)
day_of_week = my_birthdate.strftime('%A')
print(day_of_week)  # Output: Saturday
```

9. When will you be (or when were you) 10,000 days old?

```python
from datetime import date, timedelta

my_birthdate = date(1990, 10, 20)
days_to_add = 10000
future_date = my_birthdate + timedelta(days=days_to_add)
print(future_date)  # Output: 2018-02-05
```
