20. To match a number with commas for every three digits, you can use the following regular expression: 

```python
^\d{1,3}(,\d{3})*$
```

Explanation:
- `^` matches the start of the string
- `\d{1,3}` matches one to three digits
- `(,\d{3})*` matches zero or more groups consisting of a comma followed by exactly three digits
- `$` matches the end of the string
